## 데이터 삭제

```python
import pandas as pd

df = pd.read_csv('./namsan.csv', encoding='EUC-KR', low_memory=False)
```

```python
df.head()
```

```python
# 컬럼들
print(df.columns)

# 마스킹 생성
mask = df.columns != 'Unnamed: 13'
print(mask)

# 슬라이싱
ns_df = df.loc[:, mask]
ns_df.head(3)
```

```python
# 컬럼 삭제한 df 리턴 (비파괴)
df.drop(['부가기호', 'Unnamed: 13'], axis=1)

# 리턴말고 원본에서 삭제하기 (inplace 추가로 원본데이터를 바꿈) (파괴)
df.drop(['부가기호', 'Unnamed: 13', '주제분류번호', ], axis=1, inplace=True)
```

```python
df.head(3)
```

```python
# Col 중에 하나라도 NaN(비어있음) 이면 그 Col 삭제
book_df = df.dropna(axis=1)
book_df.head(3)
# col 의 모든 데이터가 NaN이면 해당 col 삭제
book_df = df.dropna(axis=1, how='all')
book_df.head(3)
```
